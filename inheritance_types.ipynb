## Multiple Inheritance

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def pay_details(self):
        return f"{self.name} earns {self.salary}"


class Coder:
    def __init__(self, language):
        self.language = language

    def code_details(self):
        return f"Codes in {self.language}"


class Developer(Employee, Coder):
    def __init__(self, name, salary, language, level):
        Employee.__init__(self, name, salary)
        Coder.__init__(self, language)
        self.level = level

    def summary(self):
        return f"{self.pay_details()} | {self.code_details()} | Level: {self.level}"
```

```python
dev = Developer("Abdullah", 150000, "Python", "Senior")
print(dev.summary())
```

## Multilevel Inheritance

```python
class Person:
    def __init__(self, name, cnic):
        self.name = name
        self.cnic = cnic

    def identity(self):
        return f"{self.name} ({self.cnic})"


class Employee2(Person):
    def __init__(self, name, cnic, employee_id):
        super().__init__(name, cnic)
        self.employee_id = employee_id

    def work_info(self):
        return f"{self.identity()} - ID: {self.employee_id}"


class Manager(Employee2):
    def __init__(self, name, cnic, employee_id, team_size):
        super().__init__(name, cnic, employee_id)
        self.team_size = team_size

    def manager_info(self):
        return f"{self.work_info()} - Manages {self.team_size} people"
```

```python
mgr = Manager("Usman Khalid", "37401-1234567-8", "EMP-220", 12)
print(mgr.manager_info())
```

## Hierarchical Inheritance

```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    def info(self):
        return f"{self.brand} {self.model}"
class Car(Vehicle):
    def __init__(self, brand, model, doors):
        super().__init__(brand, model)
        self.doors = doors
    def details(self):
        return f"{self.info()} - {self.doors} doors"
class Bike(Vehicle):
    def __init__(self, brand, model, cc):
        super().__init__(brand, model)
        self.cc = cc
    def details(self):
        return f"{self.info()} - {self.cc}cc"
class Truck(Vehicle):
    def __init__(self, brand, model, capacity_tons):
        super().__init__(brand, model)
        self.capacity_tons = capacity_tons
    def details(self):
        return f"{self.info()} - Capacity {self.capacity_tons} tons"
```

```python
vehicles = [
    Car("Toyota", "Corolla", 4),
    Bike("Honda", "CG125", 125),
    Truck("Hino", "500", 10)
]
for v in vehicles:
    print(v.details())
```

## Hybrid Inheritance

```python
class Account:
    def __init__(self, account_no, balance):
        self.account_no = account_no
        self.balance = balance
    def account_info(self):
        return f"Account {self.account_no}: {self.balance}"
class SavingsAccount(Account):
    def __init__(self, account_no, balance, interest_rate):
        super().__init__(account_no, balance)
        self.interest_rate = interest_rate
    def add_interest(self):
        self.balance += self.balance * self.interest_rate
        return self.balance
class CurrentAccount(Account):
    def __init__(self, account_no, balance, overdraft_limit):
        super().__init__(account_no, balance)
        self.overdraft_limit = overdraft_limit
    def available_balance(self):
        return self.balance + self.overdraft_limit
class PremiumCustomer:
    def __init__(self, relationship_manager):
        self.relationship_manager = relationship_manager
    def rm_info(self):
        return f"RM assigned: {self.relationship_manager}"

class PremiumSavingsAccount(SavingsAccount, PremiumCustomer):
    def __init__(self, account_no, balance, interest_rate, relationship_manager):
        SavingsAccount.__init__(self, account_no, balance, interest_rate)
        PremiumCustomer.__init__(self, relationship_manager)
    def full_summary(self):
        return f"{self.account_info()} | Interest: {self.interest_rate} | {self.rm_info()}"
```

```python
psa = PremiumSavingsAccount("PK001", 500000, 0.05, "Saba Hussain")
print(psa.full_summary())
psa.add_interest()
print(psa.account_info())
```
