#  Sustainable Development Goals (SDGs)

![SDG Icons Grid](https://globalgoalscms.co.uk/wp-content/uploads/2021/11/4a65871b85e9ab7423192498e7ce4f0b759b4448_grid.jpg)

The **Sustainable Development Goals (SDGs)**, also known as the **Global Goals**, were adopted by all United Nations Member States in 2015 as part of the **2030 Agenda for Sustainable Development**. They consist of **17 interconnected goals** that serve as a shared blueprint for peace and prosperity for people and the planet, addressing challenges such as poverty, inequality, climate change, environmental degradation, peace, and justice.

## Goal 1: No Poverty

Aims to eradicate extreme poverty in all its forms by 2030, ensuring social protection, equal rights to economic resources, and resilience for the poor and vulnerable against climate and economic shocks.

## Goal 2: Zero Hunger

Seeks to end hunger and all forms of malnutrition by 2030, promote sustainable agriculture, double agricultural productivity, and ensure everyone has access to safe, nutritious, and sufficient food.

## Goal 3: Good Health and Well-being

Focuses on ensuring healthy lives and promoting well-being for all ages, reducing maternal and child mortality, combating infectious diseases, and achieving universal health coverage.

## Goal 4: Quality Education

Aims to ensure inclusive, equitable, and quality education and promote lifelong learning opportunities for all, including free primary and secondary education and equal access to technical and higher education.

## Goal 5: Gender Equality

Seeks to achieve gender equality and empower all women and girls by ending discrimination and violence, ensuring equal opportunities in leadership, and guaranteeing equal rights to economic resources.

## Goal 6: Clean Water and Sanitation

Aims to ensure availability and sustainable management of water and sanitation for all, including universal access to safe drinking water and improved water quality through reduced pollution.

## Goal 7: Affordable and Clean Energy

Focuses on ensuring access to affordable, reliable, sustainable, and modern energy for all, while substantially increasing the share of renewable energy in the global energy mix.

## Goal 8: Decent Work and Economic Growth

Promotes sustained, inclusive, and sustainable economic growth, full and productive employment, and decent work for all, while eradicating forced labor and child labor.

## Goal 9: Industry, Innovation and Infrastructure

Aims to build resilient infrastructure, promote inclusive and sustainable industrialization, and foster innovation, including expanding access to technology and research.

## Goal 10: Reduced Inequalities

Seeks to reduce inequality within and among countries by empowering marginalized groups, promoting social, economic, and political inclusion, and ensuring equal opportunity regardless of background.

## Goal 11: Sustainable Cities and Communities

Aims to make cities and human settlements inclusive, safe, resilient, and sustainable through access to affordable housing, safe transport, and sustainable urban planning.

## Goal 12: Responsible Consumption and Production

Focuses on ensuring sustainable consumption and production patterns, including efficient use of natural resources, reducing food waste, and promoting environmentally sound management of chemicals and waste.

## Goal 13: Climate Action

Calls for urgent action to combat climate change and its impacts by strengthening resilience, integrating climate measures into national policies, and improving climate-related education and awareness.

## Goal 14: Life Below Water

Aims to conserve and sustainably use oceans, seas, and marine resources by reducing marine pollution, protecting marine ecosystems, and regulating overfishing.

## Goal 15: Life on Land

Seeks to protect, restore, and promote sustainable use of terrestrial ecosystems, sustainably manage forests, combat desertification, and halt biodiversity loss.

## Goal 16: Peace, Justice and Strong Institutions

Promotes peaceful and inclusive societies, access to justice for all, and building effective, accountable, and transparent institutions at all levels.

## Goal 17: Partnerships for the Goals

Aims to strengthen global partnerships and means of implementation, including finance, technology, capacity-building, and trade, to support the achievement of all the SDGs.
